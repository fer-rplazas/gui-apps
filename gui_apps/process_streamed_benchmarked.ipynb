```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib widget
```

```python
df = pd.read_csv("stream_benchmark.log", sep=";", header=0)

```

```python
df.head()
```

```python
from sklearn.metrics import balanced_accuracy_score
```

```python
balanced_accuracy_score(df['CNN'].values > 0.5, df['y_true']), balanced_accuracy_score(
    df['SVM'].values, df['y_true']
),

```

```python
fig, ax = plt.subplots()

ax.plot(df['CNN'].values * 0.95 + 0.05 / 2, alpha=0.5, label='CNN')
ax.plot(df['SVM'].values * 0.9 + 0.1 / 2, alpha=0.5, label='SVM')
ax.plot(df['y_true'].values, alpha=0.5, label='Acc/EMG')

ax.legend()
```
